```python
from sklearn import datasets
from sklearn.decomposition import PCA
from matplotlib import pyplot as plt
```

```python
datasets.load_iris().keys()
```

```python
label_names = datasets.load_iris().target_names
colors = ['red', 'yellow', 'green']
```

```python
samples, labels = datasets.load_iris(return_X_y=True)
print(samples.shape)
print(labels.shape)
```

```python
pca = PCA(n_components=2)
samples_pca = pca.fit(samples).transform(samples)
```

```python
for label, label_name, color in zip([0, 1, 2], label_names, colors):
    plt.scatter(samples_pca[labels==label, 0], samples_pca[labels==label, 1], label=label_name)

plt.legend(loc='lower right')
```
